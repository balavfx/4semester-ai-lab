```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sc
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data['flower_type']=iris.target
```

```python
iris_data
```

```python
iris_x=iris_data.iloc[:,[0,1,2,3]].values
```

```python
iris_y=iris_data.iloc[:,4].values
```

```python
plt.figure(figsize=(20,7))
plt.scatter(iris_x[iris_y==0,0],iris_x[iris_y==0,1],s=100,c='indigo',label='set-1')
plt.scatter(iris_x[iris_y==1,0],iris_x[iris_y==1,1],s=100,c='maroon',label='set-2')
plt.scatter(iris_x[iris_y==2,0],iris_x[iris_y==2,1],s=100,c='lime',label='set-3')
plt.legend()
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.title('Dentrograms')
```

```python
sc.dendrogram(sc.linkage(iris_x,method='ward'))
```
