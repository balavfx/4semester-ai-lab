# Creating a data

```python
movies = ['Money heist','avengers','Spiderman','fast&furious','Harryporter']
percentage = [90,80,85,95,99]
```

# Import libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# scatter plot

```python
plt.scatter(movies,percentage)
```

# line plot 

```python
plt.plot(movies,percentage,linestyle='dashdot')
```

# pie chart

```python
plt.pie(percentage,labels=movies)
```

```python
plt.bar(movies,percentage)
```

# working with inbuilt dataset

```python
tip=sb.load_dataset('tips')
```

```python
tip.tail()
```

```python
tip.head()
```

# Relationship plot

```python
sb.relplot(x='total_bill',y='tip',data=tip,col='time',hue='day',style='day',kind='scatter')
```

# Categorical box plot

```python
sb.catplot(x='day',y='total_bill',kind='box',data=tip)
```

# Regression plot

```python
sb.regplot(x='total_bill',y='tip',data=tip)     
```
